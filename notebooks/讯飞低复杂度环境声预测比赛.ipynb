## 数据预处理
把 *.wav 转化为 label_*.wav - label

```python
import os
list = os.listdir(data_path)
X_train = []
y_train = []
for d in list:
    label = d[0] # 类别
    for f in os.listdir(data_path+'\\'+d):
        X = '/'+d+'/'+f
        X_train.append(X)
        y_train.append(label)
# X_train, y_train
```

```python
df = pd.DataFrame(columns=('X', 'y'))
df['X'] = X_train
df['y'] = y_train
# df.to_csv('../data/sound_meta.csv')
df.tail()
```
